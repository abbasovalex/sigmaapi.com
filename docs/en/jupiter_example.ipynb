```python
import requests
```

```python
# add site to queue 
your_url = "https://github.com/abbasovalex/sigmaapi.com"
response = requests.get("https://sigmaapi.com/v1/add?uri=%s" % your_url)
if response.status_code == 201:
    data = response.json()
    request_id = data["requestId"]
    
print(request_id)
```

```python
# get source of the site from queue by request_id
response = requests.get("https://sigmaapi.com/v1/get?requestId=%s" % request_id)
if response.status_code == 200:
    data = response.json()
print(data)
```
